# Calibration of the venturimeter
### Analysis of the experimental data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
#Define data
Q_sr = np.array([[0, 0.000224594, 0.000357251, 0.000476615, 0.000543374, 0.000626952, 0.000762401, 0.00077683]]).transpose()
delta_I = np.array([[0, 0.762514637, 1.195826074, 1.58113883, 1.805151359, 2.08120844, 2.503140884, 2.556783694]]).transpose()
```

```python
plt.scatter(delta_I, Q_sr, s=50, c="None", marker="o", edgecolors="black")
xmin, xmax, ymin, ymax = plt.axis([0, 3, 0, 0.001])
plt.show()
```

```python
regr = LinearRegression()
regr.fit(delta_I, Q_sr)
print("Theta 0:", regr.intercept_[0])
print("Theta 1:", regr.coef_[0][0])
```

```python
plt.scatter(delta_I, Q_sr, s=50, c="None", marker="o", edgecolors="black")
plt.plot(delta_I, regr.predict(delta_I), c="black")
xmin, xmax, ymin, ymax = plt.axis([0, 3, 0, 0.001])
plt.show()
```
